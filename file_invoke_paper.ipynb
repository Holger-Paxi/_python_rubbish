```python
import pandas as pd
import glob
# import math
```

```python
references = pd.read_excel('~/Downloads/Papers/References_Literature_Review_v2.xlsx')
references.sort_values(by=['Author', 'Title', 'Publication Year'], inplace=True, ignore_index=True)
references
```

```python
references.columns.to_list()
```

```python
invoke = references['File Attachments']
invoke
```

```python
invoke_bool = invoke.notna()
invoke_bool
```

```python
invoke_bool.isna().sum()
```

```python
ata = {}
for ind_1 in invoke.index:
    # a1 = ''
    if invoke_bool[ind_1] == True:
        a2 = invoke[ind_1].split('; ')
        a1 = ''
        for ind_2 in a2:
            if (ind_2 != ' ') & (ind_2[-4:] == '.pdf') & ('\'' in ind_2):
                a1 = a1 + "Invoke-Item \"{}\";".format(ind_2)
                ata[ind_1] = a1
            elif (ind_2 != ' ') & (ind_2[-4:] == '.pdf') & (('“'or'”') in ind_2):
                a1 = a1 + "Invoke-Item '{}';".format(ind_2)
                ata[ind_1] = a1
            elif (ind_2 != ' ') & (ind_2[-4:] == '.pdf'):
                a1 = a1 + "Invoke-Item '{}';".format(ind_2)
                ata[ind_1] = a1
    else:
        ata[ind_1] = None
        # ata[ind_1] = a1

references['Invoke'] = pd.Series(ata)
references.head()
```
