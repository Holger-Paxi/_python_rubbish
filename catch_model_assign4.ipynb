```python
files = pd.DataFrame(data=glob.glob(pathname='../49255_catchment_modelling_spring_2023/assignment_4/*'), columns=['file_path'])
files.sort_values(by='file_path', inplace=True)
files.reset_index(drop=True, inplace=True)
files
```

```python
extensions = files.file_path.str.split(pat='.').str[-1].drop_duplicates()
extensions.reset_index(drop=True, inplace=True)
extensions
```

```python
extensions_to_not_open = extensions.copy()
# extensions_to_not_open.drop(labels=[3, 4, 5, 7, 8, 9], inplace=True)
extensions_to_not_open.drop(labels=[0], inplace=True)
extensions_to_not_open.reset_index(drop=True, inplace=True)
extensions_to_not_open
```

```python
files['extensions'] = files.file_path.str.split(pat='.').str[-1]
files
```

```python
files = files.set_index(keys='extensions').drop(index=extensions_to_not_open).reset_index()
files
```

```python
files['surname_name'] = files.file_path.str.split(pat='/').str[3].str.split(pat='_', n=1).str[0]
files.drop(columns='extensions', inplace=True)
files
```

```python
files = pd.DataFrame(data=files.groupby(by='surname_name'), columns=['surname_name', 'data'])
# files.drop(index=files[files.student == 'submissions.zip'].index, inplace=True)
# files.drop(index=files[files.student == 'text.md'].index, inplace=True)
# files.drop(index=files[files.student == 'text.csv'].index, inplace=True)
files.reset_index(drop=True, inplace=True)
files
```

```python
def asdasd(arg_df):
    arg_df.drop(columns='surname_name', inplace=True)
    arg_df.file_path = '/home/holger/Downloads/_rubbish' + arg_df.file_path.str.split(pat='.', n=2).str[2]
    arg_df.file_path = arg_df.file_path.apply(func=lambda arg: "xdg-open '{}'".format(arg))
    # arg_df.file_path = arg_df.file_path.apply(func=lambda arg: "nohup xdg-open -u '{}' >temp.out 2>&1 &".format(arg))
    # arg_df['extension'] = arg_df.file_path.str.split(pat='.').str[-1]
    return arg_df
```

```python
# files['files'] = files.data.apply(func=lambda arg: asdasd(arg))
files.data = files.data.apply(func=lambda arg: asdasd(arg))
files
```

```python
def qweqwe(arg_df):
    return '; '.join([ind for ind in arg_df.file_path])
```

```python
files.data = files.data.apply(lambda arg: qweqwe(arg))
files
```

```python
files = stu_names.merge(right=files, how='outer', on='surname_name')
files
```
