```python
import pandas as pd
```

```python
papers = pd.read_csv('~/Downloads/Papers.csv')
papers.sort_values(by=['Author', 'Title', 'Publication Year'], inplace=True, ignore_index=True)
papers
```

```python
papers.columns.to_list()
```

```python
path_papers = papers['File Attachments']
path_papers
```

```python
path_papers.isna().sum()
```

```python
path_papers_bool = path_papers.notna()
path_papers_bool
```

```python
ata = {}
for ind_1 in path_papers.index:
    # a1 = ''
    if path_papers_bool[ind_1] == True:
        a2 = path_papers[ind_1].split('; ')
        a1 = ''
        for ind_2 in a2:
            if (ind_2 != ' ') & (ind_2[-4:] == '.pdf') & ('\'' in ind_2):
                a1 = a1 + "open \"{}\";".format(ind_2)
                ata[ind_1] = a1
            elif (ind_2 != ' ') & (ind_2[-4:] == '.pdf') & (('“'or'”') in ind_2):
                a1 = a1 + "open '{}';".format(ind_2)
                ata[ind_1] = a1
            elif (ind_2 != ' ') & (ind_2[-4:] == '.pdf'):
                a1 = a1 + "open '{}';".format(ind_2)
                ata[ind_1] = a1
    else:
        ata[ind_1] = None
        # ata[ind_1] = a1

papers['Open'] = pd.Series(ata)
papers['Open']
```
